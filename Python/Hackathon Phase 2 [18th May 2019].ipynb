#### Problem statement:
* Using the Matplotlib library, generate a Pie Chart to categorise student marks data 
in a CSV file into Distinction, First Class, Second Class, Third Class and Failed Categories(based on percentages). 
Explode the Distinction category.


```python
import random
def genmarksdata(n,filename):
    with open(filename,'w') as marksfile:
        for i in range(0,n):
            marks=random.randint(0,100)
            marksfile.write(str(marks)+ '\n')
    return   

genmarksdata(2000,'DataFiles/marksdata.csv')
```

```python
import random
import matplotlib.pyplot as plt
def readmarksfromfile(filename):
    marks_li=[]
    with open(filename,'r') as f:
        for x in f:
            marks_li.append(int(x))
    writemarksintofile('DataFiles/marksdata.csv',marks_li)
    
def writemarksintofile(filename,arr):
    dist=0
    first=0
    second=0
    third=0
    fail=0
 
    with open(filename,'w') as reportFile:
        for i in range(0,len(arr)):
            if arr[i]>80:
                dist=dist+1
            elif(arr[i]>=60 and arr[i]<80):
                first=first+1
            elif(arr[i]>=50 and arr[i]<60):
                second=second+1
            elif(arr[i]>=40 and arr[i]<50):
                third=third+1
            else:
                fail=fail+1
    labels = 'distinction', 'first class', 'second class', 'third class','fail'
    sizes=[int(dist),int(first),int(second),int(third),int(fail)]
    explode = (0.1, 0, 0, 0, 0)
    plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    plt.axis('equal')  

    plt.show()
readmarksfromfile('DataFiles/marksdata.csv')
```
