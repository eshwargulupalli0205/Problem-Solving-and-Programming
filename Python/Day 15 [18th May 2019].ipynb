## Problem 2:-
#### Problem Statement:-
Plot a line graph on time vs distance 

```python
import numpy as np
t = np.arange(0.0,2.0,0.1)
v=1+np.sin(2*np.pi*t)
plt.xlabel('time')
plt.ylabel('voltage')
plt.grid(True)
plt.plot(t,v)
plt.savefig('figures/time-voltage.png')
plt.show()
```

## Problem 4:-
#### Problem Statement:-
Generate a "Scatter plot" for 50 random numbers

```python
n=50
x=np.random.rand(n)   # Generating 50 random numbers
y=np.random.rand(n)   # Generating 50 random numbers
colorlist=np.random.rand(n) # Generating 50 random colors
arealist=np.pi*(15*np.random.rand(50))**2 # Generating 50 random areas
plt.scatter(x,y,s=arealist,c=colorlist,alpha=1)
plt.savefig('figures/Scatter plot of 50 random numbers')
#plt.show()
```
